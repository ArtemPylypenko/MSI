```python
def bubble_sort(arr):
    n = len(arr)
    swaps_count = 0
    
    for i in range(n - 1):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                # Swap the elements
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swaps_count += 1
    
    return arr, swaps_count

# Введення списку в рядок через пробіл
input_str = input("Введіть елементи списку через пробіл: ")
input_list = list(map(int, input_str.split()))

# Сортування та отримання відсортованого списку та кількості перестановок
sorted_list, swaps = bubble_sort(input_list)

print("Відсортований список:", sorted_list)
print("Кількість перестановок:", swaps)
```
